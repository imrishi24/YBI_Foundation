# Mileage Prediction -  Regression Analysis

## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Data

```python
df = pd.read_csv(r'https://github.com/YBI-Foundation/Dataset/raw/main/MPG.csv')
```

```python
df.head()
```

```python
df.nunique()
```

## Data Preprocessing

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

## Drop Missing Values

```python
df = df.dropna()
```

```python
df.info()
```

## Data Visualization

```python
sns.pairplot(df, x_vars= ['displacement', 'horsepower', 'weight', 'acceleration', 'mpg'])
```

```python
sns.regplot(x = 'displacement', y = 'mpg', data = df)
```

 ## Define Target Variable y and Feature X

```python
df.columns
```

```python
y=df['mpg']
```

```python
y.shape
```

```python
x = df[['displacement', 'horsepower', 'weight', 'acceleration']]
```

```python
x.shape
```

```python
x
```

## Scaling Data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x=sc.fit_transform(x)
```

```python
x
```

```python
pd.DataFrame(x).describe()
```

## Train Test Split Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=182529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

## Predict Test Data

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

## Model Accuracy

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
```

```python
x_train2 = poly.fit_transform(x_train)
```

```python
x_test2 = poly.fit_transform(x_test)
```

```python
lr.fit(x_train2,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred_poly = lr.predict(x_test2)
```

## Model Accuracy

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred_poly)
```

```python
mean_absolute_percentage_error(y_test, y_pred_poly)
```

```python
r2_score(y_test, y_pred_poly)
```
